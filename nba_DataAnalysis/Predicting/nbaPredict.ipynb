```python
import pandas as pd
```

```python
df = pd.read_csv("nba_games.csv", index_col=0)
```

```python
df
```

```python
df = df.sort_values("date")
df = df.reset_index(drop=True) # resets the index to the sorted list, but drops the aditional column with the old indexes
del df["mp.1"]
del df["mp_opp.1"]
del df["index_opp"]

# need to remove some games that have buggy box scores
remove = list(range(20056, 20079))
remove.append(21044)
remove.append(21045)
df = df.drop(remove, axis='index').copy()
df = df.reset_index(drop=True)
```

```python
def add_target(team):
    team["target"] = team["won"].shift(-1) # takes in a team's games and makes a column if the team won their next game
    return team

df = df.groupby("team", group_keys = False).apply(add_target) # if group_keys is not false there will be unnecessary indexes
```

```python
# After splitting data frame using groupby, we can use data frame for each team
df[df["team"] == "GSW"]
```

```python
df["target"][pd.isnull(df["target"])] = 2 # N/As become 2s
```

```python
df["target"] = df["target"].astype(int, errors = "ignore") # convert target from boolean to int without errors
```
