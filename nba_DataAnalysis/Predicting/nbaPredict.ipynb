```python
import pandas as pd
```

```python
df = pd.read_csv("nba_games_2.csv", index_col=0)
```

```python
df = df.sort_values("date")
df = df.reset_index(drop=True) # resets the index to the sorted list, but drops the aditional column with the old indexes
del df["mp.1"]
del df["mp_opp.1"]
del df["index_opp"]
```

```python
def add_target(team):
    team["target"] = team["won"].shift(-1) # takes in a team's games and makes a column if the team won their next game
    return team

df = df.groupby("team", group_keys = False).apply(add_target) # if group_keys is not false there will be unnecessary indexes
```

```python
# After splitting data frame using groupby, we can use data frame for each team
df[df["team"] == "GSW"]
```

```python
df["target"][pd.isnull(df["target"])] = 2 # N/As become 2s
```

```python
df["target"] = df["target"].astype(int, errors = "ignore") # convert target from boolean to int without errors
```

```python
from sklearn.model_selection import TimeSeriesSplit # helps make predictions taking time into account
from sklearn.feature_selection import SequentialFeatureSelector # removes data from a dataset to improve speed
from sklearn.linear_model import RidgeClassifier # complex correlation model

rr = RidgeClassifier(alpha=1) # https://towardsdatascience.com/the-power-of-ridge-regression-4281852a64d6

# we can't do cross validation because you can't use future data to predict past data.
split = TimeSeriesSplit(n_splits=3) # split data into 3 sections?

# will choose which columns to train model on
sfs = SequentialFeatureSelector(rr, n_features_to_select=30, direction="forward", cv=split)
```

```python
# preparing columns for scaling, so the data can be easily trained on.
# scaling means to fit data into a range
removed_columns = ["season", "date", "won", "target", "team", "team_opp"]
selected_columns = df.columns[~df.columns.isin(removed_columns)]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[selected_columns] = scaler.fit_transform(df[selected_columns])
df
```

```python
# select 30 best features for predicting target
sfs.fit(df[selected_columns], df["target"])
```

```python
# get the best columns from the SequentialFeatureSelector
predictors = list(selected_columns[sfs.get_support()]) # get support will be true if that columns was chosen by the sfs
predictors
```

```python
# machine learning!
def backtest(data, model, predictors, start = 2, step = 1): # data is the data frame
    all_predictions = [] # list of dataframes, where each dataframe is the predictions for a single season
    seasons = sorted(data["season"].unique())

    for i in range(start, len(seasons), step):
        season = seasons[i]
        
        train = data[data["season"]<season]
        test = data[data["season"]==season]

        # so I had an error, where I was training the model on the test targets.
        model.fit(train[predictors], train["target"]) # .fit() is the method to train the model.

        # once model is created then make predictions
        preds = model.predict(test[predictors]) # this will try and guess whether they won their next game.
        preds = pd.Series(preds, index = test.index)

        combined = pd.concat([test["target"], preds], axis = 1)
        combined.columns = ["actual", "prediction"]

        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(df, rr, predictors)
predictions
```

```python
# accuracy of the model
from sklearn.metrics import accuracy_score

accuracy_score(predictions["actual"], predictions["prediction"])
```
